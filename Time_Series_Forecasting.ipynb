```python
#29. WHITE NOISE

#It is completely unpredictable
#In forecasting, white noise refers to a type of random or unpredictable variation that exists in a time series data. It is a fundamental concept in time series analysis and forecasting and plays a crucial role in understanding the properties of data and building accurate forecasting models.

#Characteristics of White Noise:

#Mean: White noise has a constant mean of zero. This means that, on average, the series does not exhibit any upward or downward trend.
#Variance: The variance of white noise is constant. It means that the magnitude of fluctuations in the data remains the same over time.
#Autocorrelation: White noise has no autocorrelation, which means that there is no relationship between the current observation and previous observations. In other words, each data point is independent of the others.
#Frequency Distribution: The distribution of white noise is typically Gaussian (normal), but it can also follow other distributions.
#How White Noise Affects Forecasting:
#White noise is a critical concept in forecasting for several reasons:
#Identifying Signal vs. Noise: When analyzing a time series, it is essential to identify patterns that may represent actual signals in the data (e.g., trends, seasonality, etc.) and distinguish them from random fluctuations or noise.
#White noise represents the random component that cannot be predicted or modeled.
#Model Evaluation: In forecasting, the accuracy of a forecasting model is often evaluated by comparing its performance against the performance of a white noise model. If a forecasting model cannot outperform a white noise model,
#it indicates that the model is not adding any value and may not be useful for predictions.
#Residual Analysis: White noise is also crucial in the context of model residuals. Residuals are the differences between the actual data and the forecasted values from a model. If the residuals display patterns or correlations
#(i.e., they are not white noise), it suggests that the model is not capturing all the relevant information in the data.
#Forecast Uncertainty: Understanding white noise helps in estimating the uncertainty or randomness associated with forecasting. It allows forecasters to assess the range of possible outcomes and the confidence level of the predictions.
#In practice, many time series forecasting methods aim to identify patterns and signals while minimizing the effect of white noise. Models like ARIMA (AutoRegressive Integrated Moving Average) and exponential smoothing are designed to
#capture patterns while handling random fluctuations.
#Overall, white noise is a crucial aspect of time series forecasting, and recognizing its presence is vital for building accurate and meaningful forecasting models.
```

```python
#AUTO-REGRESSION IS A LINEAR REGRESSION MODEL
# The prediction of output is based on a linear combination of inputs
# Equation of SIMPLE LINEAR REGRESSION MODEL
#  = B0+B1*X1
#where  = prediction
#B0 = coeffecients found by optimizing the model on training data
#B1 = coeffecients found by optimizing the model on training data
#X1 = input value

#STEPS:
#1.Feed a set of X & Y values to the model
#2.Model tries to fit a set of straight line through the values. Finds B0 & B1 by minimizing MSE(mean squared error)
#3.Use B0 & B1 to predict the future values of Y

#NOTE: For AUTO-REGRESSION model, lag values of some variables are used as input variable
```

##SARIMA, which stands for Seasonal Autoregressive Integrated Moving Average,
 is a time series forecasting model that builds upon the concepts of ARIMA (Autoregressive Integrated Moving Average) by adding a seasonal component. It is designed to capture both the autoregressive (AR), moving average (MA), and seasonal patterns present in a time series data.

Here's a breakdown of each component of the SARIMA model:

Autoregressive (AR) Component: This part of the model considers the relationship between the current value of the time series and its past values. The AR component represents how the current value of the series depends on its own past values.

Integrated (I) Component: This component involves differencing the time series to make it stationary. Stationarity is important because many time series models, including ARIMA and SARIMA, assume that the data has constant statistical properties over time. Differencing helps remove trend and seasonality, making the series more suitable for modeling.

Moving Average (MA) Component: This part of the model considers the relationship between the current value and past prediction errors (residuals). The MA component captures the impact of past unexpected shocks on the current value of the time series.

Seasonal Component: This is the main addition in SARIMA compared to ARIMA. Seasonal patterns often occur in time series data, such as sales data with yearly or quarterly cycles. The seasonal component captures these repeating patterns over specific intervals, like days, months, or quarters.

Now, let's break down the notation used in SARIMA(p, d, q) x (P, D, Q, s):

p: The autoregressive order, which represents the number of lagged observations included in the model.
d: The degree of differencing, which is the number of times the series needs to be differenced to achieve stationarity.
q: The moving average order, which represents the number of lagged forecast errors (residuals) included in the model.
P: The seasonal autoregressive order, similar to 'p' but for the seasonal component.
D: The seasonal differencing order, similar to 'd' but for the seasonal component.
Q: The seasonal moving average order, similar to 'q' but for the seasonal component.
s: The number of time steps in each seasonal period (e.g., 12 for monthly data with yearly seasonality).
The steps to build a SARIMA model typically involve the following:

Data Preparation: Prepare your time series data by examining and addressing trends, seasonality, and outliers. If necessary, apply differencing to achieve stationarity.

Order Selection: Determine the values of p, d, q, P, D, Q, and s using techniques like grid search, AIC (Akaike Information Criterion), or BIC (Bayesian Information Criterion).

Model Fitting: Fit the SARIMA model to your training data using the selected orders.

Forecasting: Use the fitted model to make future forecasts based on historical data.

Evaluation: Evaluate the model's performance using appropriate metrics such as Mean Squared Error (MSE), Mean Absolute Error (MAE), or others relevant to your specific problem.

SARIMA models are effective for time series forecasting when data exhibit both autoregressive and seasonal patterns. However, they may require careful tuning of the model orders, which can be a bit complex.

##47. STATIONARY TIME SERIES

A stationary time series refers to a time series where
###statistical properties remain constant over time.
In other words, the properties of the time series, such as mean, variance, and autocorrelation,
###do not exhibit any
###significant trend,
###seasonality, or
###irregular patterns that change with time.

There are two main types of stationarity:

#Strict Stationarity:
A time series is strictly stationary if the joint distribution of any set of its time points is invariant under shifts in time. This means that the statistical properties of the series do not change at different time points.

#Weak Stationarity (also known as Covariance Stationarity):
 A time series is weakly stationary if its mean, variance, and autocorrelation structure remain constant over time. It allows for some shifts in the distribution as long as the overall properties remain constant.

Stationarity is an important concept in time series analysis because many time series analysis methods, like autoregressive integrated moving average (ARIMA) models, assume stationarity to work effectively. When a time series is not stationary, it might be necessary to transform or preprocess the data to make it stationary before applying certain analyses or models. This can involve differencing the series to remove trends and seasonality, among other techniques.

We have been mentioning about some models which are applicable on only time series data,
###which does not have trend and seasonality, but there is a special term for such series.
Such series are called
###stationary Series.

Well, technically, a stationary time series is one whose statistical properties, such as
###mean variance, autocorrelation, etc., are all constant over time.

For most business and economic activities, the Time series data is rarely stationary.

For example, if you look at the mean of the monthly sales of a company for 2010 and then for 2019, most probably the means will be different.

That company would have either grown or declined in sales performance over time.

So since the mean is changing over time, this time series is not stationary.

###And we can not apply models such as ar or arima to forecast it.

###So to make a time series stationary, we just need to use differencing , as we have shown you earlier.

###We will use differencing to remove trend and seasonality both.

And for most purposes, we will be able to use our models on the resulting series.

###Else you can simply use models such as SARIMA or SARIMAX, which handles seasonality and trend.
